```python
# Ejercicio 17
datos = {'Mes':['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'],
         'Ventas':[10000, 15200, 14300, 25800, 2560, 12780, 32500, 23624, 12000, 15858, 23250, 11630],
         'Gastos':[6500, 6500, 6500, 7850, 6500, 8250, 6500, 9252, 6500, 10250, 12450, 13000],
         'Beneficio':[3500,8700,7800,17950,-3940,4530,26000,14372,5500,5608,10800,-1370],
         'Cobrado':['S', 'N', 'N', 'S', 'S', 'N', 'S', 'S', 'S', 'N', 'S', 'N']}
df_datos = pd.DataFrame(datos)
df_datos["Dividendo"] = df_datos["Beneficio"]
df_datos
```

```python
# Ejercicio 18
df_datos = df_datos.assign(Dividendo = df_datos["Dividendo"] * 0.1)
df_datos
```

```python
# Ejercicio 19
df_datos["Dividendo"] = df_datos["Dividendo"].map(lambda x: 0 if x < 0 else x)
df_datos
```

```python
# Ejercicio 20
def calcular_porcentaje_div(elemento):
    porcentaje_div = (elemento["Dividendo"]/elemento["Ventas"])*100
    return round(porcentaje_div, 2)
    
df_datos["Porcentaje_Dividendo"] = df_datos.apply(calcular_porcentaje_div, axis = 1)
df_datos
```

```python
# Ejercicio 21
campos_numericos = []
diccionario_datos = df_datos.dtypes.to_dict()

for clave, valor in diccionario_datos.items():
    if valor != object:
        campos_numericos.append(clave)

df_datos_2 = df_datos.drop(columns=campos_numericos)
df_datos_2
```

```python
# Ejercicio 22
campos_df = []
diccionario_datos = df_datos.dtypes.to_dict()
df_num = df_datos

for clave, valor in diccionario_datos.items():
    if valor == object and clave != "Cobrado":
        campos_df.append(clave)

df_num = df_num.drop(columns=campos_df)
tabla_agrup = df_num.groupby("Cobrado").mean().round(2)
tabla_agrup
```
