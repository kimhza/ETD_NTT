```python
# Ejercicio 6
for col in df_report.columns:
    q1 = df_report[col].quantile(0.25)
    q3 = df_report[col].quantile(0.75) 
    IQR = q3 - q1

    # Cálculo de limites
    lim_inf = round((q1 - 1.5 * IQR), 2)
    lim_sup = round((q3 + 1.5 * IQR), 2)

    # Cáluclo de valores por fuera de los rangos inferiores y superiores
    outlier_inf = df_report[col] < lim_inf
    outlier_sup = df_report[col] > lim_sup

    outliers = df_report.loc[outlier_inf | outlier_sup, col] 
    print(f"* {col} -> [Limite inferior: {lim_inf}, Limite superior: {lim_sup}]")

    if len(outliers) > 0:
        print(f"\tValores atípicos para {col}: \n{outliers}")
    else:
        print(f"\tSin valores atípicos para {col}")
```

```python
# Ejercicio 8
def calcular_freq(cols):
    for c in cols:
        col = df_report[c].value_counts(normalize=True)
        lista_vals_freq = []
        conteo = 0
        print(f"{'-'*70}\n* {c}")
        
        for clave, valor in col.items():
            valor = round(valor*100, 1)
            if valor > 5:
                conteo += 1
            else:
                lista_vals_freq.append(clave)   
        
        if conteo == len(col.values):
            print(f"Todos los registros de '{c}' tienen una freq. superior al 5%")
        else:
            print(f"Existen {conteo} registros de '{c}' que tienen una freq. inferior al 5%:\n{lista_vals_freq}")

cols = ["counted", "population", "percentage"]
calcular_freq(cols)
```
